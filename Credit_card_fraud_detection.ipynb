```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# Loading a Dataset to a Pandas DataFrame
credit_card_data = pd.read_csv('/content/creditcard.csv')
```

```python
# First 5 rows of the dataset
credit_card_data.head()
```

```python
# Last 5 rows of the dataset
credit_card_data.tail()
```

```python
# Dataset informations
credit_card_data.info()
```

```python
# Checking the numbers of missing values in each column
credit_card_data.isnull().sum()
```

Handling the missing values

```python
# Drop the very last row of the Dataframe
credit_card_data.drop(credit_card_data.tail(1).index,inplace=True)
```

```python
# Last 5 rows of the dataset
credit_card_data.tail()
```

```python
# Checking the numbers of missing values in each column
credit_card_data.isnull().sum()
```

```python
# Distribution of legit and fraudulent transactions
 # ( 0=Normal transactions, 1=Fraudulent transaction)
credit_card_data['Class'].value_counts()
```

This data is Highly Unbalanced

0=Normal transactions
1=Fraudulent transaction

```python
# Seperating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# Statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# Compare the values for the both transactions
credit_card_data.groupby('Class').mean()
```

Under Sampling

Build a sample dataset containing simple distribution of normal transaction and fraudulent transacton

**Number of Fraudulent transaction --> 492**

```python
legit_sample = legit.sample(n=492)
```

```python
# Concatenating two DataFrames
new_dataset = pd.concat([legit_sample, fraud], axis = 0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class']. value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

Splitting the data into Features and Targets

```python
x=new_dataset.drop(columns='Class', axis=1)
y=new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

Splitting the data into Training data and Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(x, y, test_size=0.2, stratify=y, random_state=2)
```

```python
print(x.shape, X_train.shape, X_test.shape)
```

Model Training

**Logistic Regression**

```python
model = LogisticRegression()
```

```python
# Training the LogisticRegression model
model.fit(X_train, Y_train)
```

Model Evaluation

> Accuracy Score



```python
# Accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction, Y_train)
```

```python
print('Accuracy on Training data : ', training_data_accuracy)
```

```python
# Accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction, Y_test)
```

```python
print('Accuracy on Test data : ', test_data_accuracy)
```
